```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load your dataset
df = pd.read_csv(r"D:\student placement prediction cgd\data\placementdata.csv")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Student Placement Data Analysis', fontsize=16)

# 1. Placement Status Distribution
sns.countplot(data=df, x='PlacementStatus', palette=['red', 'green'], ax=axes[0,0])
axes[0,0].set_title('Placement Status Distribution')

# 2. Extracurricular Activities vs Placement
sns.countplot(data=df, x='ExtracurricularActivities', hue='PlacementStatus', ax=axes[0,1])
axes[0,1].set_title('Extracurricular Activities vs Placement')

# 3. Placement Training vs Placement
sns.countplot(data=df, x='PlacementTraining', hue='PlacementStatus', ax=axes[0,2])
axes[0,2].set_title('Placement Training vs Placement')

# 4. CGPA by Placement
sns.boxplot(data=df, x='PlacementStatus', y='CGPA', ax=axes[1,0], palette="Set2")
axes[1,0].set_title('CGPA by Placement Status')

# 5. Internships by Placement (better as barplot than boxplot)
sns.countplot(data=df, x='Internships', hue='PlacementStatus', ax=axes[1,1])
axes[1,1].set_title('Internships vs Placement')

# 6. Aptitude Test Score by Placement
sns.violinplot(data=df, x='PlacementStatus', y='AptitudeTestScore', ax=axes[1,2], palette="Set3")
axes[1,2].set_title('Aptitude Test Score by Placement')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
